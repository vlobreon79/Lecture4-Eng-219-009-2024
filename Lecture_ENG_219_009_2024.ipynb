```python
def maximum(value1, value2, value3):
   """Return the maximum of three values."""
   max_value = value1
   if value2 > max_value:
    max_value = value2
   if value3 > max_value:
    max_value = value3
   return max_value
```

```python
print(maximum(12, 27, 36))
print(maximum(12.3, 45.6, 9.7))
print(maximum('yellow', 'red', 'orange'))
```

**Exercise 04.3 (functions)**

```python
def area(x0, y0, x1, y1, x2, y2):
  """Compute the area of a triangle given the coordinates of its vertices."""
  return 0.5 * abs(x0 * (y1 - y2) + x1 * (y2 - y0) + x2 * (y0 - y1))
```

### `area(x0, y0, x1, y1, x2, y2)` Function

This function calculates the area of a triangle using the coordinates of its three vertices $(x_0, y_0)$, $(x_1, y_1)$, and $(x_2, y_2)$. The formula used is based on the determinant method:

$$\text{Area} = \frac{1}{2} |x_0(y_1 - y_2) + x_1(y_2 - y_0) + x_2(y_0 - y_1)|$$

```python
# Test cases for the area function

# Test 1: A right-angled triangle (base=4, height=3)
calculated_area_1 = area(0, 0, 4, 0, 0, 3)
print(f"Area of triangle (0,0), (4,0), (0,3): {calculated_area_1}")
assert calculated_area_1 == 6.0

# Test 2: Another simple triangle
# Vertices: (1,1), (5,1), (3,4)
# Base = 4 (from x=1 to x=5), Height = 3 (from y=1 to y=4)
calculated_area_2 = area(1, 1, 5, 1, 3, 4)
print(f"Area of triangle (1,1), (5,1), (3,4): {calculated_area_2}")
assert calculated_area_2 == 6.0

# Test 3: Collinear points (should result in area 0)
# Vertices: (0,0), (1,1), (2,2)
calculated_area_3 = area(0, 0, 1, 1, 2, 2)
print(f"Area of triangle (0,0), (1,1), (2,2): {calculated_area_3}")
assert calculated_area_3 == 0.0

# Test 4: Triangle with negative coordinates
# Vertices: (-2,-2), (2,-2), (0,2)
# Base = 4 (from x=-2 to x=2), Height = 4 (from y=-2 to y=2)
calculated_area_4 = area(-2, -2, 2, -2, 0, 2)
print(f"Area of triangle (-2,-2), (2,-2), (0,2): {calculated_area_4}")
assert calculated_area_4 == 8.0

# New test case provided by the user
x0, y0 = 0.0, 0.0
x1, y1 = 0.0, 2.0
x2, y2 = 3.0, 0.0
A = area(x0, y0, x1, y1, x2, y2)
import math
assert math.isclose(A, 3.0)

print("All test cases passed!")
```

**Exercise 04.4 (recursion)**

```python
import math

def factorial(n):
  """Compute the factorial of a non-negative integer recursively."""
  if n == 0 or n == 1:
    return 1
  else:
    return n * factorial(n - 1)
```

```python
print("Reference factorial:", math.factorial(5))
print("Factorial of 5:", factorial(5))

print("Reference value of factorial of 5:", math.factorial(5))

## tests ##
assert factorial(0) == 1
assert factorial(1) == 1
assert factorial(2) == 2
assert factorial(5) == 120

assert factorial(32) == math.factorial(32)

print("All factorial tests passed!")
```

```python
import sympy
from sympy import symbols, Function, dsolve, Eq, Derivative, solve, sympify

x = symbols('x')
y = Function('y')

print("Solving a second-order linear ODE of the form: a*y''(x) + b*y'(x) + c*y(x) = f(x)")

try:
    a = float(input("Enter coefficient 'a' (for y''): "))
    b = float(input("Enter coefficient 'b' (for y'): "))
    c = float(input("Enter coefficient 'c' (for y): "))
    f_str = input("Enter the non-homogeneous term f(x) (e.g., 'sin(x)', 'exp(x)', '0'): ")
    f_expr = sympify(f_str)
except (ValueError, SyntaxError) as e:
    print(f"Invalid input: {e}. Please enter numeric values for coefficients and a valid sympy expression for f(x).")
    raise

homogeneous_ode_eq = Eq(a*Derivative(y(x), x, 2) + b*Derivative(y(x), x) + c*y(x), 0)
homogeneous_sol = dsolve(homogeneous_ode_eq, y(x))
print(f"\nHomogeneous Solution (y_h): {homogeneous_sol.rhs}")

full_ode_eq = Eq(a*Derivative(y(x), x, 2) + b*Derivative(y(x), x) + c*y(x), f_expr)
general_sol = dsolve(full_ode_eq, y(x))
print(f"General Solution (y_g): {general_sol.rhs}")

C1, C2 = symbols('C1 C2')
if f_expr == 0:
    particular_sol_display = sympify(0)
else:
    print("\nNote: SymPy's dsolve combines homogeneous and particular solutions. ")
    print("The particular solution (y_p) is the part of the general solution that remains")
    print("if the homogeneous solution (y_h) is subtracted (disregarding constants C1, C2).")

print("\nNow, let's apply initial conditions to find the specific solution.")

try:
    x0 = float(input("Enter x0 (e.g., '0'): "))
    y0 = float(input("Enter y(x0) (e.g., '1'): "))
    yp0 = float(input("Enter y'(x0) (e.g., '0'): "))
except ValueError:
    print("Invalid input for initial conditions. Please enter numeric values.")
    raise

sol_expr = general_sol.rhs
sol_deriv = Derivative(sol_expr, x).doit()

eq1 = Eq(sol_expr.subs(x, x0), y0)
eq2 = Eq(sol_deriv.subs(x, x0), yp0)

# FIX: Correctly extract the integration constants from the general solution's free symbols
sympy_constants = sorted([s for s in general_sol.free_symbols if s.name.startswith('C')], key=lambda s: s.name)

if len(sympy_constants) == 2:
    C1_sym, C2_sym = sympy_constants[0], sympy_constants[1]
    constant_values = solve((eq1, eq2), (C1_sym, C2_sym))

    if constant_values:
        specific_sol_expr = sol_expr.subs(constant_values)
        print(f"\nConstants found: {constant_values}")
        print(f"Specific Solution: y(x) = {specific_sol_expr}")
    else:
        print("\nCould not find unique constants for the given initial conditions.")
else:
    print(f"\nExpected 2 constants (C1, C2) but found {len(sympy_constants)}. Cannot apply initial conditions.")

print("\nCalculations complete.")
```
